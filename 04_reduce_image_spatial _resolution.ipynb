```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def block_reducing(image, block_size):
    h, w = image.shape
    new_img = image.copy()
    
    for i in range(0, h - block_size + 1, block_size):
        for j in range (0, w - block_size +1, block_size):
            block = image[i:i+block_size,j:j+block_size]
            avg = int(np.mean(block))
            new_img[i:i+block_size, j:j+block_size] = avg
    return new_img
```

```python
# Load grayscale image
img = cv2.imread('image_03.jpeg')
# Convert to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Original RGB image
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')

# Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Apply block avaraging for 3x3, 5x5, 7x7
block_size = [3,5,7]

plt.figure(figsize=(20, 5))
for idx, bsize in enumerate(block_size):
    result = block_reducing(gray_image, bsize)
    plt.subplot(1, len(block_size)+1, idx+2)
    plt.imshow(result, cmap='gray')
    plt.title(f"{bsize}x{bsize} Block Avarage")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
