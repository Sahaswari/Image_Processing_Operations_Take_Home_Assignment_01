```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
import cv2

from tkinter import filedialog
from skimage import io
from skimage import data, filters, util
from skimage.filters import gaussian
from scipy.ndimage import median_filter
from skimage import exposure
from skimage import io, color
```

```python
#upload a image
root = tk.Tk()
root.withdraw()
file_path = filedialog.askopenfilename(title="Select an image file", filetypes=[("Image files", "*.png;*.jpg;*.jpeg;*.bmp")])
org_image = io.imread(file_path)
plt.imshow(org_image)
```

```python
# Apply average filters of different sizes
kernel_sizes = [3, 10, 20]

for k in kernel_sizes:
   # filtered_image = gaussian(org_image, sigma=1, channel_axis=-1)
  #  plt.Subplot(1,2,1);plt.title("filterd Image");plt.show();
    blurred_image = cv2.blur(org_image, (k, k), 0);
    output_filename = f"blurred_image_{k}.png"
    io.imsave(output_filename, blurred_image)
    #Disply the results
    plt.subplot(1,2,1);plt.title("Orignal Image - "+f"{k}x{k}.png");plt.imshow(org_image);
    plt.subplot(1,2,2);plt.title("filterd Image -"+ f"{k}x{k}.png");plt.imshow(blurred_image);
    plt.show()
    
    
```
