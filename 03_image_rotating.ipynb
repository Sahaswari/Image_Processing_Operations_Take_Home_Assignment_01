```python
import cv2
import matplotlib.pyplot as plt
```

```python
def rotate_image(image, angle):
    (h,w) = image.shape[:2]
    center = (w // 2, h // 2)
    rot_mat = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, rot_mat, (w, h))
    return rotated
```

```python
image = cv2.imread('image_04.jpeg')
plt.title('Original Image'); plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize=(10, 10))
rotated_45 = rotate_image(cv2.imread('image_04.jpeg'), 45)
rotated_90 = rotate_image(cv2.imread('image_04.jpeg'), 90)
plt.subplot(1,3,1); plt.title('Original Image'); plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(1,3,2); plt.title('45 Rotated  Image'); plt.imshow(cv2.cvtColor(rotated_45, cv2.COLOR_BGR2RGB))
plt.subplot(1,3,3); plt.title('90 Rotated Image'); plt.imshow(cv2.cvtColor(rotated_90, cv2.COLOR_BGR2RGB))
```
