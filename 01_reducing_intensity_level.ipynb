```python
#Import the libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog 
```

```python
#Load the Image
Tk().withdraw()
file_path = filedialog.askopenfilename(title="Select an Image")
image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
#Reduce the intensity level
def reduce_intensity_levels(img, levels):
    factor = 256 // levels
    reduced = (img // factor) * factor
    return reduced

for levels in [2,4,8]:
    reduced_image = reduce_intensity_levels(image, levels)
    plt.imshow(reduced_image, cmap="gray")
    plt.title(f"Reduced to {levels} Levels")
    plt.axis('off')
    plt.show()
```
